```python
#labeling된 파일들 모두 new_X와 new_y로 변환
new_X=[]
new_y=[]
import numpy as np
import cv2

for t in range(10):
#해당 Label파일에 cutting된 사진 저장 
    os.chdir(f"./{t}")
    print(os.listdir())
    files=os.listdir()        
    #for j in range(10):
    for idx, file in enumerate(files):
        fname, ext = os.path.splitext(file)
    
        #디렉토리에있는 파일중 jpg,png파일만 읽어들임
        if ext in ['.jpg','.png']:
            path=os.getcwd()
            print(path,file)
            img1= cv2.imread(file)
            #한픽셀 반전해서 담기
            bit=[]
            for i in range(28):
                for j in range(28):
                    a=0
                    a= 255-img1[i][j][0]
                    #비트담기
                    bit.append(a)
            
            new_X.append(bit)
            new_y.append(t)

            #모두 잘라낸뒤 resize파일로 돌아가기
    os.chdir("../")
```
